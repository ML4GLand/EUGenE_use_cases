# Set-up

```python
# Imports
import os
import sys
import numpy as np
import h5py
import xarray as xr

sys.path.append("/cellar/users/aklie/projects/ML4GLand/use_cases/ResidualBind/bin")
from dataload import load_rnacompete_data, get_experiment_names, find_experiment_index
```

```python
# Paths
input_path = "/cellar/users/aklie/data/ml4gland/pubs/koo21_gia/rnacompete2013.h5"
outdir_path = "/cellar/useras/aklie/data/ml4gland/pubs/koo21_gia/"
```

# Load data

```python
exp_accession = get_experiment_names(input_path)
exp_index = [find_experiment_index(input_path, i) for i in exp_accession]
```

```python
train, valid, test = load_rnacompete_data(
    file_path=input_path,
    ss_type="seq",
    normalization="log_norm",
    rbp_index=None,
    dataset_name=None
)
```

```python
train.keys(), valid.keys(), test.keys()
```

# Create SeqDatas

## Train dataset

```python
# Complete dataset
ohe_seq = xr.DataArray(train["inputs"], dims=("_sequence", "_length", "_ohe"))
targets = xr.DataArray(train["targets"], dims=("_sequence", "_target"))
rbp_id = xr.DataArray(exp_accession, dims=("_target"))
rbp_index = xr.DataArray(exp_index, dims=("_target"))
sdata = xr.Dataset({"inputs": ohe_seq, "targets": targets, "rbp_id": rbp_id, "rbp_index": rbp_index})
sdata.to_zarr(os.path.join(outdir_path, "log_norm", "rnacompete2013_train.zarr"), mode="w")
sdata
```

## Validation dataset

```python
# Do the same thing for validation and test sets
ohe_seq = xr.DataArray(valid["inputs"], dims=("_sequence", "_length", "_ohe"))
targets = xr.DataArray(valid["targets"], dims=("_sequence", "_target"))
rbp_id = xr.DataArray(exp_accession, dims=("_target"))
rbp_index = xr.DataArray(exp_index, dims=("_target"))
sdata = xr.Dataset({"inputs": ohe_seq, "targets": targets, "rbp_id": rbp_id, "rbp_index": rbp_index})
sdata.to_zarr(os.path.join(outdir_path, "log_norm", "rnacompete2013_valid.zarr"), mode="w")
sdata
```

## Test dataset

```python
ohe_seq = xr.DataArray(test["inputs"], dims=("_sequence", "_length", "_ohe"))
targets = xr.DataArray(test["targets"], dims=("_sequence", "_target"))
rbp_id = xr.DataArray(exp_accession, dims=("_target"))
rbp_index = xr.DataArray(exp_index, dims=("_target"))
sdata = xr.Dataset({"inputs": ohe_seq, "targets": targets, "rbp_id": rbp_id, "rbp_index": rbp_index})
sdata.to_zarr(os.path.join(outdir_path, "log_norm", "rnacompete2013_test.zarr"), mode="w")
sdata
```
